```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a funo f(x)
def f(x):
    return (x - 10**2)**2 + ((6557 / x) - 10**2)**2 - 2 * (np.sqrt(6557) - 10**2)**2

# Gerando valores de x com alta resoluo entre 1 e 300 para capturar a curva de subida
x = np.linspace(1, 100, 100000)  # Usando 10.000 pontos para alta granularidade

# Calculando f(x) para cada valor de x
y = f(x)

# Plotando o grfico
plt.figure(figsize=(20, 20))
plt.plot(x, y, label="f(x)", color='blue')
plt.title("Plot da funo f(x) com ajuste de intervalo")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=0.5)  # Linha horizontal em y=0
plt.axvline(0, color='black', linewidth=0.5)  # Linha vertical em x=0
plt.legend()
plt.grid(True)
plt.show()

```
